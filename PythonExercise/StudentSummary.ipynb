```python
import pandas as pd
```

```python
df=pd.read_csv('student_random_data_with_summary.csv')
print(df)
```

```python
avg_marks=df['Marks'].mean()
max_marks=df['Marks'].max()
toppers = df[df["Marks"] == max_marks]["Name"].tolist()
print(avg_marks)
print(max_marks)
print(toppers)
```

```python
summary_data = {
    "Average Marks": [avg_marks],
    "Highest Marks": [max_marks],
    "Topper(s)": [", ".join(toppers)]
}

summary_df = pd.DataFrame(summary_data)
summary_df.to_csv("student_summary.csv", index=False)

df1=pd.read_csv("student_summary.csv")

print(df1)
```
